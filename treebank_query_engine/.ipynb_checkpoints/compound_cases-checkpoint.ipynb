```python
import sqlite3
import sys
import glob
import re
import os
import numpy as np
from wxconv import WXC


conn = sqlite3.connect('Treebank_English.db')
print ('Opened database successfully');
cursor = conn.cursor()
```

```python
cursor.execute('SELECT a.sid,a.wid,a.word,a.rel,a.pos_UD,b.sid,b.wid,b.word,b.rel,b.pos_UD from Tword a join Tword b on a.sid=b.sid and a.filename=b.filename where b.parent=a.wid and b.rel="compound";')
myresult = cursor.fetchall()
n=len(myresult)

f=open("E_compound_cases.dat",'w')

for i in range(n):
    #print(myresult[i])
    
    if myresult[i][4]!='VERB' and myresult[i][9]!='VERB':
        print("sentenceid: "+myresult[i][0]+"\tparent: "+myresult[i][2]+"\tPOS: "+myresult[i][4]+"\tchild: "+myresult[i][7]+"\tPOS: "+myresult[i][9])
        f.write("sentenceid: "+myresult[i][0]+"\tparent: "+myresult[i][2]+"\tPOS: "+myresult[i][4]+"\tchild: "+myresult[i][7]+"\tPOS: "+myresult[i][9])
f.close()  
    
```
